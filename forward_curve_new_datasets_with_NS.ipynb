## DATA grid-based approach

```python
# Size of the data
train_size = 40000; # Size of the training set
test_size = 4000; # Size of the test set
calibration_size = 4000;
param_in = 7; # Input size to the neural network

# Two dimensional output sizes (the image)
K_vector = np.array([31.6, 31.8, 32.0, 32.2, 32.4, 32.6, 32.8, 33.0, 33.2])
tau_vector = np.array([1/12, 2/12, 3/12, 4/12, 5/12, 6/12, 1]) 

N1 = len(K_vector)
N2 = len(tau_vector) 

l = 1/12 # 1 month of delivery, fixed
# NOTE: T_1 = tau, assumption for all the experiments
```

```python
params_range = create_features_linspace(N = train_size + test_size,
                                          v_min = [0.2, 0.5, 8, 34.2, -1.5, 0.2, 4.5], # [a, b, k, a0, a1, a2, a3]
                                          v_max = [0.5, 0.8, 9, 34.7, -1.0, 1.2, 5.0]) # [a, b, k, a0, a1, a2, a3]

price = np.empty((train_size + test_size, N1, N2))
#vol = np.empty((train_size + test_size, N2))
for i in range(train_size + test_size):
    print("i=", i)
    model_features = params_range[i,:3]
    #s = uni_volatility(model_features = model_features)
    NS_param = [params_range[i,3].item(), params_range[i,4].item(), params_range[i,5].item(), params_range[i,6].item()]
    price_BS = black_scholes_price(NS_param = NS_param)
    for j in np.arange(N1):
        K = K_vector[j]
        for k in np.arange(N2):
            price[i,j,k] = price_BS(option_features = np.array([K, tau_vector[k], tau_vector[k], l]),
                                    model_features = model_features)
            #if (price[i,j,k]<1e-2):
            #    print("--- price = ", price[i,j,k], " for (j, k) = (", j, ",", k, ")")
            #vol[i,k] = s(tau = tau_vector[k], T_1 = tau_vector[k], l = l)
```

```python
quanti = 0
for k in np.arange(N2):
    for j in np.arange(N1):
        quanti+= np.sum(price[:,j,k]<1e-2)
print([quanti, N1*N2*train_size, quanti/(N1*N2*train_size)])
```

```python
np.savetxt('params_gridbased_with_NS.dat', params_range)
#np.savetxt('vol_gridbased_with_NS.dat', vol)
np.savetxt('price_gridbased_with_NS.dat', np.reshape(price, newshape = (train_size+test_size, N1*N2), order ='F'))
```

## DATA pointwise approach

```python
# Size of the data
train_size = 60000; # Size of the training set
test_size = 6000; # Size of the test set
calibration_size = test_size;
param_in = 9; # Input size to the neural network
parameter_cal = 7; # Size of training variables

l = 1/12 # 1 month of delivery, fixed
# NOTE: T_1 = tau, assumption for all the experiments
```

```python
params_range = create_features_linspace(N = train_size + test_size,
                                          v_min = [31.6, 1/12,  0.2, 0.5, 8, 34.2, -1.5, 0.2, 4.5], # [a, b, k, a0, a1, a2, a3]
                                          v_max = [33.2, 1, 0.5, 0.8, 9, 34.7, -1.0, 1.2, 5.0]) # [a, b, k, a0, a1, a2, a3]

price = np.empty((train_size + test_size,1))
#vol = np.empty((train_size + test_size,1))
for i in range(train_size + test_size):
    print("i=", i)
    model_features = params_range[i,2:5]
    #s = uni_volatility(model_features = model_features)
    NS_param = [params_range[i,5].item(), params_range[i,6].item(), params_range[i,7].item(), params_range[i,8].item()]
    price_BS = black_scholes_price(NS_param = NS_param)
    K = params_range[i,0]
    tau = params_range[i,1]
    price[i] = price_BS(option_features = np.array([K, tau, tau, l]),
                                    model_features = model_features)
    #if (price[i]<1e-2):
    #    print("--- price = ", price[i], " for (K, tau) = (", K, ",", tau, ")")
    #vol[i] = s(tau = tau, T_1 = tau, l = l)
```
